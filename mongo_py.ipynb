```python
# Connect to Mongo
client = MongoClient('localhost', 27017)

db = client["nyc_taxis"]
```

```python
## number of trips per VendorID

pipeline = [
    {"$unwind": "$id_trip"},
    {"$group": {"_id": "$vendor_id", "myCount": { "$sum": 1 }}}
]

count_vendor = time_coll.aggregate(pipeline)
```
